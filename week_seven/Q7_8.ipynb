```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = 500
mu = np.array([5.0, 0.0, 10.0])

r = np.array([
        [  3.40, -2.75, -2.00],
        [ -2.75,  5.50,  1.50],
        [ -2.00,  1.50,  1.25]
    ])

y = np.random.multivariate_normal(mu, r, size=n)
```

```python
# Plot various projections of the samples.
plt.subplot(2,2,1)
plt.plot(y[:,0], y[:,1], 'b.')
plt.plot(mu[0], mu[1], 'ro')
plt.ylabel('y[1]')
plt.axis('equal')
plt.grid(True)

plt.subplot(2,2,3)
plt.plot(y[:,0], y[:,2], 'b.')
plt.plot(mu[0], mu[2], 'ro')
plt.xlabel('y[0]')
plt.ylabel('y[2]')
plt.axis('equal')
plt.grid(True)

plt.subplot(2,2,4)
plt.plot(y[:,1], y[:,2], 'b.')
plt.plot(mu[1], mu[2], 'ro')
plt.xlabel('y[1]')
plt.axis('equal')
plt.grid(True)

plt.show()
```

```python
n = 500
mu = np.array([2, 2])

r = np.array([
        [5, -0.01],
        [-0.01, 5],
    ])

y = np.random.multivariate_normal(mu, r, size=n)
np.corrcoef(y[:,0], y[:,1])[0,1]
```

```python
plt.plot(y[:,0], y[:,1], 'b.')
plt.ylabel('y[1]')
plt.axis('equal')
plt.grid(True)
plt.title('r = %.3f' % np.corrcoef(y[:,0], y[:,1])[0,1])
```
