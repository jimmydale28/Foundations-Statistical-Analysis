```python
import numpy as np
import random
import matplotlib.pyplot as plt
from scipy import stats
%matplotlib inline
```

## Part A
#### Men make $5,000 more than women

```python
men = [(random.randint(1,9)*10000) for _ in range(100)]
women = [(men[i]-5000) for i in range(len(men))]
slope, intercept, r_value, p_value, std_err = stats.linregress(men,women)
```

```python
plt.plot(men, women)
plt.title(r_value)
plt.show()
```

#### Correlation is 1

## Part B
#### Men make 25% more than woman

```python
women = [(random.randint(1,9)*10000) for _ in range(100)]
men = [(women[i]*1.25) for i in range(len(women))]
print(women[0:10])
print(men[0:10])
```

```python
slope, intercept, r_value, p_value, std_err = stats.linregress(men,women)
```

```python
plt.plot(men, women)
plt.title(r_value)
plt.show()
```

#### Correlation is 1
